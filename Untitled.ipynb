```python
import akshare as ak

# 获取单个股票数据
data = ak.stock_zh_a_daily(symbol="sh600519")

# 按日期排序（防止数据倒序）
data = data.sort_index()

# 取最近 60 个交易日
recent = data.tail(60)

# 计算动量
momentum = (recent["close"].iloc[-1] - recent["close"].iloc[0]) / recent["close"].iloc[0]
print("动量：", momentum)
```

```python
import akshare as ak
import pandas as pd

# 获取沪深300成分股
stocks = ak.index_stock_cons(symbol="000300")["品种代码"].tolist()

momentum_list = []
for code in stocks:
    try:
        # Akshare 需要前缀区分沪深
        prefix = "sh" if code.startswith("6") else "sz"
        data = ak.stock_zh_a_daily(symbol=f"{prefix}{code}")
        data = data.sort_index()

        if len(data) >= 60:
            recent = data.tail(60)
            mom = (recent["close"].iloc[-1] - recent["close"].iloc[0]) / recent["close"].iloc[0]
            momentum_list.append({"code": code, "momentum": mom})
    except Exception as e:
        print(f"{code} 失败: {e}")

df_momentum = pd.DataFrame(momentum_list).sort_values("momentum", ascending=False)
print(df_momentum.head())
```
